```python
import numpy as np
import pandas as pd
from random import randint
```

```python
def rank_swap(p,input_path):
    
    data = input_path
    
    df = pd.read_csv(data)
    
    def conv(row):
        currency_dict = {"BGN": 1.9558, "NZD": 1.6805, "ILS": 4.0448, "RUB": 72.2099, "CAD": 1.5075, "USD": 1.1218,
                     "PHP": 58.125, "CHF": 1.1437, "ZAR": 16.0224, "AUD": 1.5911, "JPY": 124.93, "TRY": 6.6913,
                     "HKD": 8.8007, "MYR": 4.6314, "THB": 35.802, "HRK": 7.413, "NOK": 9.6678, "IDR": 15953.68,
                     "DKK": 7.4646, "CZK": 25.659, "HUF": 322.97, "GBP": 0.86248, "MXN": 21.2829, "KRW": 1308.01,
                     "ISK": 136.2, "SGD": 1.5263, "BRL": 4.405, "PLN": 4.2868, "INR": 78.0615, "RON": 4.7596,
                     "CNY": 7.5541, "SEK": 10.635}
        return row['amount'] / (currency_dict[row['currencycode']]*100)

    df['amount_eur'] = df.apply(lambda x: conv(x), axis=1)
    
    sortby_amt_eur = df.sort_values('amount_eur')
    
    sortedbyamtindex=sortby_amt_eur.index.to_list()
    
    param = p
    
    interval = int(len(df.index)*(p/100))
    value_list=[]
    for i in range(len(sortedbyamtindex)):
        if i in value_list:
            continue
        else:    
            value = randint(0,290381)
            while(abs(i-value)>interval):
                value = randint(0,290381)
            value_list.append(value)
            sortedbyamtindex[i], sortedbyamtindex[value] = sortedbyamtindex[value], sortedbyamtindex[i]
    
    print("halfway done")
    newdf=df.reindex(sortedbyamtindex)  
    
    amount=newdf["amount"]
    
    currencycode=newdf["currencycode"]
    
    df["amount"]=amount.values
    
    df["currencycode"]=currencycode.values
    
    df.drop(columns=["amount_eur"],inplace=True)
    
    sortby_date = df.sort_values('creationdate')
    
    sortby_date.head()
    
    sortedbydateindex=sortby_date.index.to_list()
    value_list=[]
    for i in range(len(sortedbydateindex)):
        if i in value_list:
            continue
        else:
            value = randint(0,290381)
            while(abs(i-value)>interval):
                value = randint(0,290381)
            value_list.append(value)    
            sortedbydateindex[i], sortedbydateindex[value] = sortedbydateindex[value], sortedbydateindex[i]
    
    newdf=df.reindex(sortedbydateindex)
    
    creationdate= newdf["creationdate"]
    
    df["creationdate"]=creationdate.values
    
    return df
```

```python
rank_swap=rank_swap(10,"data_for_student_case.csv")
```
